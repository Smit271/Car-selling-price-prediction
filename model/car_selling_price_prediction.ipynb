```python
# Importing Useful Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading data into dataframe

df = pd.read_csv('../data/Car_details_v3.csv')
df.head()
```

```python
# Dataset Information

df.info()
```

```python
# Droping nan values from all rows to make our dataset clear

df.dropna(inplace= True)
df.info()
```

```python
# Dropping name and year of car as not to make model too complicated

df.drop(['name', 'year', 'torque'], axis='columns', inplace = True)
df.head()
```

```python
# After Dropping nan reseting index 

df.reset_index(drop = True, inplace= True)
df.tail()
```

```python
# Checking fuel column's unique values for encoding

df.fuel.unique()
```

```python
# Encoding phase of fuel column

df['fuel'] = df['fuel'].apply(lambda x: df.fuel.unique().tolist().index(x) + 1 )
```

```python
df.tail()
```

```python
# Same as fuel do with all columns containing categorical data 

df.seller_type.unique()
```

```python
# OneHot Encoding to encode categorical column to numeric 

temp = pd.get_dummies(df['seller_type'])
df = pd.concat([df, temp], axis = 'columns')
df.drop(['seller_type'], axis = 'columns', inplace = True)
df
```

```python
# same as above 

df.transmission.unique()
```

```python
df['transmission'] = df['transmission'].apply(lambda x: df.transmission.unique().tolist().index(x))
df.tail()
```

```python
# same as above

df.owner.unique()
```

```python
# Extra (Optional)

temp = df.owner.unique().tolist()
temp.reverse()
```

```python
df['owner'] = df['owner'].apply(lambda x: temp.index(x) + 1 )
df.tail()
```

```python
# Removing kmpl and making numeric value understandable by model

df['mileage'] = df['mileage'].apply(lambda x : float(str(x).split(" ")[0]))
df.head()
```

```python
# same as mileage column removing CC from engine column

df['engine'] = df['engine'].apply(lambda x : float(str(x).split(" ")[0]))
df.tail()
```

```python
# same as all above removing bhp from max_power column

df['max_power'] = df['max_power'].apply(lambda x : float(str(x).split(" ")[0]))
df.tail()
```

```python
# Let's see Information now after preprocessing and cleaning of our data 

df.info()
```

```python
# Putting Independent and dependent in X and y

X = df.iloc[:, 1:].values
y = df.iloc[:, 0].values
```

```python
# Splitting our dataset into train and test data set to make model more effiecient

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.20 )
```

```python
# Importing All our model

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
```

```python
# Let's make 10 different model with 10 different data set with Linear_regression

cross_val_score(LinearRegression(n_jobs=-1), x_train, y_train, cv=10)
```

```python
#Let's make 10 different model with 10 different data set with DecisionTree

cross_val_score(DecisionTreeRegressor(), x_train, y_train, cv=10)
```

```python
#Let's make 10 different model with 10 different data set with RandomForest with 500 tress
# This will take time as per your computer capacity

cross_val_score(RandomForestRegressor(n_estimators= 500, n_jobs=-1), x_train, y_train, cv = 10)
```

<h3>Now we can say from all regressions model Random Forest is bit better from other so let's choose it</h3>

```python
# making model of RandomForest Regression model with 500 trees 

model = RandomForestRegressor(n_estimators= 500, n_jobs=-1)
```

```python
# Training our model with our training data

model.fit(x_train, y_train)
```

```python
# Checking testing accuracy to check if overfitting is happen or not 

model.score(x_test, y_test)
```

```python
# Training Accuracy

model.score(x_train, y_train)
```

```python
# To Deployment of our model there is need of model file so let's make it by help of joblib library

import joblib
joblib.dump(model, 'car_prediction')
```
